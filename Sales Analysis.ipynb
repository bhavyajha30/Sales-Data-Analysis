```python
# What is overall sales trend?

df["order_date"].min()
```

```python
df["order_date"].max()
```

```python
#Getting month_year from dataset
df["month_year"]=df["order_date"].apply(lambda x:x.strftime('%Y-%m'))
df["month_year"]
```

```python
#Grouping month_year:
df_trend=df.groupby("month_year")[["sales"]].sum().reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend["month_year"], df_trend["sales"])
plt.xticks(rotation='vertical', size=8)
plt.show()
```

```python
# What are the top 10 products by sales?

prod_sales=pd.DataFrame(df.groupby("product_name")["sales"].sum())
```

```python
prod_sales=prod_sales.sort_values("sales", ascending=False)
prod_sales[:10]
```

```python
# Which are the most selling products?

most_sold_prod=pd.DataFrame(df.groupby("product_name")[["quantity"]].sum())
most_sold_prod=most_sold_prod.sort_values("quantity", ascending=False)
```

```python
most_sold_prod[:10]
```

```python
# What is the most preferred ship mode?
sns.countplot(x=df["ship_mode"])
```

```python
# Which are the most profitable category and sub-category

cat_subcat_profit=pd.DataFrame(df.groupby(["category", "sub_category"])["profit"].sum())
cat_subcat_profit.sort_values(["category", "profit"], ascending=False)
```
